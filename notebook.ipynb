```python
import pandas as pd
```

```python
df = pd.read_csv("ETHUSDT-trades-2022-09-30.csv",
                names = ["id","price","qty","quoteQty","time","makerBuy","bestPrice"])

df["time"] = pd.to_datetime(df["time"], unit = "ms")
df.set_index("time", inplace = True)
```

```python
df.price.resample("1H").agg({
    "open":"first",
    "high":"max",
    "low":"min",
    "close":"last"
})
```
